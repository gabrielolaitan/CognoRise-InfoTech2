```python
# Changing the date datatype from int to datetime
file['work_year'] = pd.to_datetime(file["work_year"], format="%Y")
file['work_year'] = file['work_year'].dt.year
```

```python
# Drop the "Unnamed" column
file1 = file.drop("Unnamed: 0", axis=1)
file1
```

```python
# Create the scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data = file1 , x='employment_type', y='salary_in_usd', hue='employment_type', alpha=0.7, s=100)

# Add labels and title
plt.title('Distribution of Salaries by Employment Type', fontsize=16)
plt.xlabel('Employment Type', fontsize=12)
plt.ylabel('Salary (USD)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Employment Type', loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```
